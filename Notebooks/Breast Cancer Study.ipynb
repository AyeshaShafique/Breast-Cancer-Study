```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix
import matplotlib.pyplot as plt  
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from collections import defaultdict
from xgboost import XGBClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import KFold

data = pd.read_csv('study-breast-cancer-wisconsin.csv')
data.head()
```

```python
data.describe()
```

```python
data.shape
```

```python
#handle missing values
data= data.fillna(data.mean())
```

```python
# Shuffle the data
data = shuffle(data) 

label = data.pop("Class")

X=data
Y=label
```

```python
kf = KFold(n_splits=10, shuffle=True)
datadistO = defaultdict(int) # int() produces 0
for a in Y:
    datadistO[a] += 1
print("Origional Data Distribution", datadistO)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(50,30,40,50,30),max_iter=4200)
i=1
for train, test in kf.split(X):
    print("\n\nEvaluation on Fold # ", i)
    X_train = X.iloc[train]
    X_test =  X.iloc[test]
    y_train = Y.iloc[train]
    y_test = Y.iloc[test]
    
    #convert array into dataframe
    X_train=pd.DataFrame(X_train, columns=list(X.columns))
    y_train=pd.DataFrame(y_train)
    
    datadistTr = defaultdict(int) # int() produces 0
    for a in y_train:
        datadistTr[a] += 1
    print("\nTrain Data Distribution", datadistTr)
    datadistTe = defaultdict(int) # int() produces 0
    for a in y_test:
        datadistTe[a] += 1
    print("\nTest Data Distribution", datadistTe)
    
    mlp.fit(X_train,y_train)
    
    P_nn = mlp.predict(X_test)
    P_nnt = mlp.predict(X_train)
    
    print("\nTraining Accuracy:",metrics.accuracy_score(y_train, P_nnt))
    print("\nTest Accuracy:",metrics.accuracy_score(y_test, P_nn))
    
    cm = confusion_matrix(y_test,P_nn)
    print("\nConfusion Matrix:\n",cm)
    
    print("Training Classification Report:\n",classification_report(y_train,P_nnt))
    print("\nTesting Classification Report:\n",classification_report(y_test,P_nn))
    
    i = i + 1
```
